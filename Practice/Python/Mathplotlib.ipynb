```python
# More datasets
data1 = np.random.normal(100, 20, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
```

```python
plt.figure(figsize=(10, 7))
labels = ['A', 'B', 'C', 'D']

boxprops = {'facecolor':'lightblue', 'edgecolor':'teal', 'linewidth':2.0}
whiskerprops = {'color': 'green', 'linewidth': 1.5}
capprops = {'color': 'orange', 'linewidth': 1.5}
medianprops = {'color': 'black', 'linewidth': 2}

plt.boxplot(data,
            patch_artist=True,
            boxprops=boxprops,
            whiskerprops=whiskerprops,
            capprops=capprops,
            meanprops=medianprops,
            labels=labels)

plt.ylabel('Value')
plt.xlabel('Dataset')
plt.title('Box plot', fontsize=20)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)
data = np.random.normal(2, 3, 500)
plot = plt.boxplot(data, patch_artist=True)

color = 'green'
fill_color='lightblue'
width = 1.7

for prop in ['boxes', 'whiskers', 'means', 'medians', 'caps']:
    plt.setp(plot[prop], color=color, linewidth=width)

for prop in ['boxes']:
    plt.setp(plot[prop], facecolor=fill_color)

plt.show()
```

## Code rand about tracking time of pico code 

```python
import time
from matplotlib import pyplot

led_strip = ['green', 'blue', 'red', 'yellow']
x, y = [], []
```

```python
def strobe(delay=0.5, speed_up=0.1):
    i = 1
    slow_down = True

    start_time = time.perf_counter()
    while time.perf_counter() - start_time <= 10:
        i = i % 4
        
        print(led_strip[i])
        x.append(time.perf_counter() - start_time)
        y.append(delay)
        
        time.sleep(delay)
        if slow_down:
            if delay <= 0.1:
                slow_down = False
            else:
                delay = delay * (1 - speed_up)
                speed_up = speed_up * 0.85
        else:
            if delay >= 1:
                slow_down = True
            else:
                delay = delay * (1 + speed_up)
                speed_up = speed_up * 1.15
        i += 1


delay = 1
speed_up = 0.3  # % of current delay
strobe(delay, speed_up)

print(x)
print(y)
```

```python
pyplot.legend()
pyplot.xlabel("Input")
pyplot.ylabel("Time [s]")
```
